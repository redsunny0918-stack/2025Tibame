python群集型態
1. List: 同類型清單
[20, 50, 30, 100]
key=0 1 2 3

2. Dict(字典): 一個複雜東西
{"姓名":"Elwing", "身高":175}
key=自訂的

3. Set: 不重複集合
{"周", "王", "黃"}
key=沒有

4. tuple: 省略版的字典
(255, 255, 255)
key=0, 1, 2

```python
names = {"周", '王', "黃"}
print(names)

# 專屬功能: add
names.add("何")
print(names)
names.add("王")
print(names)

# 交集: & 我們共有的東西
names2 = {"黃", "謝", "蔡"}
print(names & names2)

# 聯集: | 所有可能
print(names | names2)

# 剪集: 我 - 交集
print(names - names2)
print(names2 - names)
```

```python
import random

def generate_ticket(low=1, high=49, count=6):
  lottery = set()
  while True:
    n = random.randint(low, high)
    lottery.add(n)
    if len(lottery) == count:
      break
  return lottery

prize = generate_ticket()
# 特別號: in
special = None
while True:
  special = random.randint(1, 49)
  if not special in prize:
    break
special = random.randint(1, 49)
print(prize)
print(special)

count = 0
target = 11
# 一直買彩劵，跟頭彩彩劵對獎，對到二獎(11) 三獎(10) 看到底要買幾張
while True:
  ticket = generate_ticket()
  count = count + 1
  score = len(ticket & prize) * 2
  if special in ticket:
    score = score + 1
  if score >= target:
    print(count, ticket)
    break
```

```python
# 如果想放到set裡或者當作字典dict key
# 這個東西一定不能可以變長/變短/修改(不可變動unmutable)
# !!!! tuple
import random
import datetime as dt

count = 0
birthdays = set()
while True:
  n = random.randint(1, 365)
  count = count + 1
  date_rand = "2025-" + str(n).zfill(3)
  date_convert = dt.datetime.strptime(date_rand, "%Y-%j")
  d = (date_convert.month, date_convert.day)
  if d in birthdays:
    print("幾個人才有同月同日生:", count)
    break
  else:
    birthdays.add(d)
    print(birthdays)


```
